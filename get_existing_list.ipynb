```python
import os
from arcgis.gis import GIS
from arcgis.features import SpatialDataFrame
from arcgis.geometry import Point
import pandas as pd
import ast

file_root = 'demo_trips'
```

```python
f_list = [f for f in os.listdir('.') if os.path.isfile(f) and f.endswith('.csv') and f.startswith(file_root)]
f_list
```

```python
df = pd.read_csv(f_list.pop(), index_col=0)

# required because SHAPE is loaded as a string, and must be a dictionary to be properly 
# recognized by the SpatialDataFrame constructor
df.SHAPE = df.SHAPE.apply(lambda v: ast.literal_eval(v)) 
```

```python
sdf = SpatialDataFrame(df)
sdf
```

```python
gis = GIS(url='https://commteamretail.maps.arcgis.com', username='jmccune_retail')
```

```python
resp = sdf.to_featurelayer('national_store_inrix_origins', gis, 'human movement')
resp
```
